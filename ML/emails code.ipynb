```python
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score
from sklearn.neighbors import KNeighborsClassifier
```

```python
df = pd.read_csv("..\dataset\emails.csv")
df
```

```python
df.isnull()
```

```python
df.isnull().sum()
```

```python
X = df.iloc[:, 1:3001].values  # Ensure X is a NumPy array
Y = df.iloc[:, -1].values  # Target variable
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.25, random_state=42)
```

```python
# C here is the regularization parameter. Here, L2 penalty is used(default). 
# It is the inverse of the strength of regularization.
# As C increases, model overfits.
# Kernel here is the radial basis function kernel.
# gamma (only used for rbf kernel) : As gamma increases, model overfits.
svc = SVC(C=1.0, kernel='rbf', gamma='auto')
svc.fit(X_train, y_train)
y_pred_svc = svc.predict(X_test)
print("Accuracy Score for SVC: ", accuracy_score(y_test, y_pred_svc))
```

```python
knn = KNeighborsClassifier(n_neighbors=7)
knn.fit(X_train, y_train)  # Fit on training data
y_pred_knn = knn.predict(X_test)
print("KNN Predictions: ", y_pred_knn)
print("KNN Accuracy Score: ", knn.score(X_test, y_test))
```
